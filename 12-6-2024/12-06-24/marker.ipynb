```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,3,6,7,10]
```

```python
#create a plot with marker
plt.plot(x,y,marker='*',linestyle='-.',color='b',markersize=10,markerfacecolor='r')

#add label as well as title

plt.title("line plot with markers")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```
