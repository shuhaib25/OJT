```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.plot(x,y1,linestyle='-',color='b',linewidth=1,marker='o',markersize=8,markerfacecolor='violet')
plt.plot(x,y2,linestyle='-.',color='black',linewidth=3,marker='o',markersize=10,markerfacecolor='green')


plt.title('customised line plot with grid')
plt.xlabel('xaxis')
plt.ylabel('yaxis')

plt.legend()

plt.grid(True)

plt.grid(color='grey',linestyle='--',linewidth=1)

plt.show()
```
