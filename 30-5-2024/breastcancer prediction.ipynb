```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('data.csv')
df.head()
```

```python
#Data exploration
df.shape
df.info()
df.select_dtypes(include='object').columns
len(df.select_dtypes(include='object').columns)
df.select_dtypes(include=['float64','int64']).columns
len(df.select_dtypes(include=['float64','int64']).columns)
```

```python
#Statistical summary
df.describe()
```

```python
#Dealing with the missing values
df.isnull().values.any()
df.isnull().values.sum()
df.columns[df.isnull().any()]
len(df.columns[df.isnull().any()])
df['Unnamed: 32'].count() #this column contains only 0/null values
df = df.drop(columns='Unnamed: 32')
df.shape
df.isnull().values.any()
```

```python
#Dealing with categorical data
df.select_dtypes(include='object').columns
df['diagnosis'].unique()
df['diagnosis'].nunique()
```

```python
#One hot encoding
data=df
df=pd.get_dummies(data=df, drop_first=True)
df.head()
(df.diagnosis_M==0).sum() #B(0) values
(df.diagnosis_M==1).sum() # M(1) values
```

```python
#Correlation matrix and heatmap
df2=df.drop(columns='diagnosis_M')
df2.head()
df2.corrwith(df['diagnosis_M']).plot.bar(
    figsize=(20,10), title='Correlated with diagnosis_M ', rot=45, grid=True)
corr= df.corr()
corr
```

```python
# heatmap
plt.figure(figsize=(20,10))
sns.heatmap(corr, annot=True)
```
