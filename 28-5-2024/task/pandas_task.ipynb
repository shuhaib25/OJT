```python
#data import and convert to data frame
import pandas as pd

df=pd.read_csv('air-pollution.csv')
print(df)
```

```python
print(df.describe())
```

```python
#clean data rows empty
cleaned_df = df.dropna(how="all")
```

```python
# Replace null values with 0
cleaned_df = cleaned_df.fillna(0)
```

```python
#remove duplicate entities
cleaned_df.drop_duplicates(inplace=True)
```

```python
# Calculate median, mean, and standard deviation for 'Nitrogen oxide (NOx)' and add them as new columns
cleaned_df['NOx_median'] = cleaned_df['Nitrogen oxide (NOx)'].median()
cleaned_df['NOx_mean'] = cleaned_df['Nitrogen oxide (NOx)'].mean()
cleaned_df['NOx_std'] = cleaned_df['Nitrogen oxide (NOx)'].std()
```

```python
cleaned_df['SO₂_median'] = cleaned_df['Sulphur dioxide (SO₂) emissions'].median()
cleaned_df['SO₂_mean'] = cleaned_df['Sulphur dioxide (SO₂) emissions'].mean()
cleaned_df['SO₂_std'] = cleaned_df['Sulphur dioxide (SO₂) emissions'].std()
```
