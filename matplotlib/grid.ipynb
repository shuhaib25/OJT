```python
import matplotlib.pyplot as plt
```

```python
#data
x = [1,2,3,4,5]
y1= [2,11,5,7,19]
y2= [1,2,3,5,8]

#create the plot 
plt.plot(x,y1,linestyle='-',color='b',linewidth=1,marker='o',markersize=8,markerfacecolor='violet')
plt.plot(x,y2,linestyle='-.',color='black',linewidth=2,marker='o',markersize=10,markerfacecolor='green')


#title and label

#create alegend method with chartbthat we have created
plt.legend()

#add the value as grid
plt.grid(True)

#costomoize the grid
plt.grid(color='grey',linestyle='-.',linewidth=1)
```
