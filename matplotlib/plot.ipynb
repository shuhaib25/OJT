```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#data set
x = [1,2,3,4,5,6,7]
y = [2,3,5,7,11,12,13]
```

```python
#create a plot for our data
plt.plot(x,y)

#customozation for the plot
#add title
plt.title('Line Plot')

#add labels
plt.xlabel("x-axis")
plt.ylabel("y-axis")
#output the plot
plt.show()
```
